```python
import h5py
import matplotlib.pyplot as plt
import os
import numpy as np
```

```python
root_dir = "/eos/project/m/ml-for-alarm-system/private/consorzio_rfx_datasets/hvptf2018/"
file_names = os.listdir(root_dir)
```

```python
f = [h5py.File(root_dir+file_name, "r") for file_name in file_names]
vm_mis_full = np.hstack([fi["ANALOG_IN"]["VM_MIS"] for fi in f])
```

```python
plt.figure(figsize=(25,10))
plt.plot(vm_mis_full[0][1::1000], ".", markersize=1)
plt.title(f"VM_MIS")
plt.savefig("vm_mis.png")
```

```python
def plot_all_signals(file, group, channel):
    plt.figure(figsize=(35,10))
    for f in file:
        plt.plot(f[group][channel][(1)],f[group][channel][(0)], alpha=0.2)
        plt.title(f"{group} - {channel}")
        #print(f"signal: {group} - {channel}, length: {len(f[group][channel][(1)])}, start time: {f[group][channel][(1)][0]}, delta: {f[group][channel][(1)][1]-f[group][channel][(1)][0]}")
    plt.savefig("vm_mis.png")
    
def plot_one_signals(file, file_nr, group, channel):
    plt.figure(figsize=(12,7))

    plt.plot(f[file_nr][group][channel][(1)],f[file_nr][group][channel][(0)])
    plt.title(f"{group} - {channel}")
    print(f"signal: {group} - {channel}, length: {len(f[file_nr][group][channel][(1)])}, start time: {f[file_nr][group][channel][(1)][0]}, delta: {f[file_nr][group][channel][(1)][1]-f[file_nr][group][channel][(1)][0]}")
```

```python
plot_all_signals(f, "ANALOG_IN", "VM_MIS")
```

```python
plot_one_signals(f, 0, "ANALOG_IN", "VM_MIS")
```

```python
def plot_group(f, group):
    for channel in f[group].keys():
        plt.figure()
        plt.plot(f[group][channel][(1)],f[group][channel][(0)])
        plt.title(f"{group} - {channel}")
        print(f"signal: {group} - {channel}, length: {len(f[group][channel][(1)])}, start time: {f[group][channel][(1)][0]}, delta: {f[group][channel][(1)][1]-f[group][channel][(1)][0]}")
```

```python
plot_group(f, "ANALOG_IN")
```

```python
plot_group(f, "ANALOG_OUT")
```
