```python
import h5py
import matplotlib.pyplot as plt
import os
from pathlib import Path
import numpy as np
```

```python
data_dir = Path().cwd().parent / "data/raw/"
file_names = os.listdir(data_dir)
file_names
```

```python
data = [h5py.File(data_dir / file_name, "r") for file_name in file_names]
vm_mis_full = np.hstack([fi["ANALOG_IN"]["VM_MIS"] for fi in data])
```

```python
plt.figure(figsize=(25,10))
plt.plot(vm_mis_full[0][1::1000], ".", markersize=1)
plt.title(f"VM_MIS")
plt.savefig("vm_mis.png")
```
